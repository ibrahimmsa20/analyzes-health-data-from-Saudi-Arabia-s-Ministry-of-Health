```
import pandas as pd
import matplotlib.pyplot as plt
import os


plt.rcParams['font.family'] = 'DejaVu Sans'
plt.style.use('seaborn-v0_8-muted')


os.makedirs('figures', exist_ok=True)

def load_data():



    df_centers = pd.DataFrame({
        'Total_Registered_Centers': [27],
        'A1C_Checked_3+_Times_%': [75],
        'Total_Patients_Registered': [330066],
        'Type_I_DM_%': [14],
        'Type_II_DM_%': [74],
        'Gestational_DM_%': [12],
        'Nurses': [349],
        'Diabetic_Foot_Specialists': [47]
    })

    df_survey = pd.DataFrame({
        'Males_Diabetes': [1293878],
        'Females_Diabetes': [1077000],
        'Males_Pre_Diabetes': [1117262],
        'Females_Pre_Diabetes': [774080],
        'Males_Pre_Diabetes_%': [17.0],
        'Females_Pre_Diabetes_%': [15.5],
        'No_Doctor_Visit_Checkup_%': [75.5],
        'Obesity_%_Males': [24.1],
        'Obesity_%_Females': [33.5],
        'High_Blood_Pressure_Males': [46.5],
        'High_Blood_Pressure_Females': [34.3],
    }, index=[0])

    return df_centers, df_survey

def save_and_show_plot(title, filename):



    plt.title(title, fontsize=14)
    plt.tight_layout()
    plt.savefig(f'figures/{filename}', dpi=300)
    plt.show()

def create_gender_diabetes_bar_chart(df_survey):



    labels = ['Males', 'Females']
    values = [df_survey['Males_Diabetes'][0], df_survey['Females_Diabetes'][0]]

    plt.figure(figsize=(8, 6))
    bars = plt.bar(labels, values, color=['steelblue', 'lightcoral'])
    plt.ylabel('Number of Diagnosed Cases', fontsize=12)
    plt.xlabel('Gender', fontsize=12)
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                 f'{int(bar.get_height()):,}', ha='center', va='bottom', fontsize=10)
    save_and_show_plot('Prevalence of Diabetes by Gender', 'diabetes_by_gender.png')

def create_diabetes_type_pie_chart(df_centers):


    labels = ['Type I Diabetes', 'Type II Diabetes', 'Gestational Diabetes']
    sizes = [df_centers['Type_I_DM_%'][0], df_centers['Type_II_DM_%'][0], df_centers['Gestational_DM_%'][0]]

    plt.figure(figsize=(8, 6))
    plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140, colors=['#66c2a5','#fc8d62','#8da0cb'])
    save_and_show_plot('Distribution of Diabetes Types Among Patients', 'diabetes_types_pie.png')

def create_obesity_gender_bar_chart(df_survey):



    labels = ['Males', 'Females']
    values = [df_survey['Obesity_%_Males'][0], df_survey['Obesity_%_Females'][0]]

    plt.figure(figsize=(8, 6))
    bars = plt.bar(labels, values, color=['steelblue', 'lightcoral'])
    plt.ylabel('Obesity Prevalence (%)', fontsize=12)
    plt.xlabel('Gender', fontsize=12)
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                 f'{bar.get_height():.1f}%', ha='center', va='bottom', fontsize=10)
    save_and_show_plot('Obesity Prevalence by Gender', 'obesity_by_gender.png')

def create_pre_diabetes_gender_bar_chart(df_survey):



    labels = ['Males', 'Females']
    values = [df_survey['Males_Pre_Diabetes_%'][0], df_survey['Females_Pre_Diabetes_%'][0]]

    plt.figure(figsize=(8, 6))
    bars = plt.bar(labels, values, color=['skyblue', 'lightpink'])
    plt.ylabel('Pre-Diabetes Prevalence (%)', fontsize=12)
    plt.xlabel('Gender', fontsize=12)
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                 f'{bar.get_height():.1f}%', ha='center', va='bottom', fontsize=10)
    save_and_show_plot('Pre-Diabetes Prevalence by Gender', 'pre_diabetes_by_gender.png')

def create_high_blood_pressure_gender_bar_chart(df_survey):



    labels = ['Males', 'Females']
    values = [df_survey['High_Blood_Pressure_Males'][0], df_survey['High_Blood_Pressure_Females'][0]]

    plt.figure(figsize=(8, 6))
    bars = plt.bar(labels, values, color=['navy', 'crimson'])
    plt.ylabel('High Blood Pressure Prevalence (%)', fontsize=12)
    plt.xlabel('Gender', fontsize=12)
    for bar in bars:
        plt.text(bar.get_x() + bar.get_width()/2, bar.get_height(),
                 f'{bar.get_height():.1f}%', ha='center', va='bottom', fontsize=10)
    save_and_show_plot('High Blood Pressure Prevalence by Gender', 'blood_pressure_by_gender.png')

def main():



    df_centers, df_survey = load_data()
    create_gender_diabetes_bar_chart(df_survey)
    create_diabetes_type_pie_chart(df_centers)
    create_obesity_gender_bar_chart(df_survey)
    create_pre_diabetes_gender_bar_chart(df_survey)
    create_high_blood_pressure_gender_bar_chart(df_survey)

if __name__ == "__main__":
    main()
```
